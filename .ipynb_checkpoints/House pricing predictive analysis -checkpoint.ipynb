# Linear Regression 

```python
#lets define the variables 'target' and 'input'
target = data_preprocessed['Log Price']
inputs = data_preprocessed.drop(['Log Price'],axis=1)
```

# Scale the  data

```python
#lets  the import the module
from sklearn.preprocessing import StandardScaler

#create a scaler object 
scalar = StandardScaler()
#Lets fit the inputs 
scalar.fit(inputs)
```

```python
#Lets tramsform the inputs 
inputs_scaled = scalar.transform(inputs)
```


# Train test split

```python
#Lets import the train test split module
from sklearn.model_selection import train_test_split

#Lets split the variables to an 80-20 split 
x_train,x_test,y_train,y_test = train_test_split(inputs_scaled,target,test_size=0.2,random_state=365)
```

# Create a  Linear regression

```python
#lets createe an object
reg = LinearRegression()
#lets fit the model
reg.fit(x_train,y_train)
```

```python
#lets check the output of the regression
y_hat = reg.predict(x_train)
```

```python
#Lets the compare the Targets(y-train) and the prediction(y_hat) on a scatter plot
plt.scatter(y_train,y_hat)
#Lets label both axis 
plt.xlabel('Targets',size=10)
plt.ylabel('Predictions',size=10)

#lets make the x axis and the y axis have the same scale 
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
#Lets plot the PDF of the Residuals and check for anomalies 
sns.distplot(y_train-y_hat)
#Lets name the plot
plt.title('Residuals PDF',size = 15)
```

```python
#lets find the R-squared 
reg.score(x_train,y_train)
```

```python
#lets calculate the adjusted R2
def adj_R2(x_train,y_train):
    r2 = reg.score(x_train,y_train)
    n = x_train.shape[0]
    p = x_train.shape[1]
    adjusted_R2 = 1-(1-r2)*(n-1)/(n-p-1)
    return adjusted_R2
adj_R2(x_train,y_train)
```

# Lets find the weights and bias 

```python
reg.intercept_
```

```python
#Lets put it in a dataframe to make it intrepretable
reg_summary = pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary['Weights'] = reg.coef_
reg_summary
```

# Testing 

```python
#lets test our model on a dataset that the Alogrithm have never seen before 
y_hat_test = reg.predict(x_test)
```

```python
#lets create a scatter plot on the test_target and the test_prediction 

plt.scatter(y_test,y_hat_test,alpha=0.1)
plt.xlabel('Target (y_test)',size=18)
plt.ylabel('prediction (y_hat)',size=18)

#lets use the same scale for both x and y axis
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
#Lets manually check  the predictions values 
df_pf = pd.DataFrame(np.exp(y_hat_test),columns = ['Predictions'])
df_pf
```

```python
#lets reset index and add the targets to the dataframe to compare with the predictions 
y_test = y_test.reset_index(drop=True)
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
#Lets add the difference between the targets and the predictions to the dataframe 
df_pf['Residual'] = df_pf['Target'] - df_pf['Predictions']
df_pf.drop('Residuals',axis=1)
```

```python
#lets see hoe far off we are from the percentage result wise 
#Lets take the absolute difference in percentage%
df_pf['Difference%'] = np.absolute(df_pf['Residual']/df_pf['Target']*100)
df_pf
```
